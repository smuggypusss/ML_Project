```python
#Dependent and independent features
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y#Target
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=5)
```

```python

x_train
```

```python
y_train
```

```python
#Standardizing the dataset,Why?
#because we use gradient descent, to reach the global minima we need to make sure that all variables have the same scale 
#to make sure that the convergence is happening quickly
from sklearn.preprocessing import StandardScaler
```

```python
#Data standardization is the process of rescaling the attributes so that they have mean as 0 and variance as 1.
scaler=StandardScaler()
```

```python
#The fit method is calculating the mean and variance of each of the features present in our data. 
#The transform method is transforming all the features using the respective mean and variance.
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Now doing linear regression
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
#Print the coefficients and the interepts
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
#On which parameters the regression has been trained
regression.get_params()
```

```python
#Prediction with test 
reg_pred=regression.predict(x_test)
```

```python
#Plot scatter for the prediction
plt.scatter(y_test,reg_pred)
```

```python
error=y_test-reg_pred 
```

```python
error
```

```python
#plot the residuals
sns.displot(error,kind="kde")
```

```python
#plot between prediction and error
plt.scatter(reg_pred,error)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
```

```python
# R square and adjusted R square
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
score
```

```python
#adjusted R square
1 - (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
